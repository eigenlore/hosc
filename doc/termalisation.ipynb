```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
term_cold=np.loadtxt('../main/data_files/thermalisation/thermalisation_cold_N512.dat')
term_hot=np.loadtxt('../main/data_files/thermalisation/thermalisation_hot_N512.dat')
```

```python
plt.plot(term_cold[:,0], term_cold[:,1], 'b', label='A freddo')
plt.plot(term_hot[:,0], term_hot[:,1], 'r', label='A caldo')
plt.xlabel('Numero di sweep')
plt.ylabel('Azione')
plt.title('Termalizzazione con N=256')
plt.legend()
plt.show()
```

L'accettanza media a freddo è: 0.593125

L'accettanza media a caldo è: 0.592891
