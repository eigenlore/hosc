```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy.optimize import curve_fit
from numpy import sqrt
from numpy import log
def f(x, a,b,c):
    return a+b*x+c*x**2
```

```python
a=2
weff=sqrt(1+a**2/4)
deltaE_esatta32=1/a*log(a*weff+1+a**2/2)
matrix_el_esatto32=1/sqrt(2*weff)
print('deltaE esatta ',deltaE_esatta32)
print('matrix_el esatto', matrix_el_esatto32)
deltaE32=np.loadtxt('../main/data_files/energy/mean_deltaE_N32.dat')
matrix_el32=np.loadtxt('../main/data_files/energy/mean_matrix_el_N32.dat')
deltaE32/=a
print(deltaE32)
print('t test deltaE', (deltaE_esatta32-deltaE32[0])/deltaE32[1])
print('t test matrix el', (matrix_el_esatto32-matrix_el32[0])/matrix_el32[1])
```

```python
x=np.array([1, 0.5, 0.25, 0.125, 0.0625])
x=x**2
y=np.array([deltaE64[0], deltaE128[0], deltaE256[0], deltaE512[0], deltaE1024[0]])
sigma_y=np.array([deltaE64[1], deltaE128[1], deltaE256[1], deltaE512[1], deltaE1024[1]])
print(sigma_y)
#x=np.array([0.0625*0.0625, 0.125*0.125, 0.25*0.25, 0.5*0.5, 1])
#y=np.array([0.999647,      0.999379,    0.997331,  0.989866, 0.962423])
#sigma_y=np.array([0.000428,      0.000292,    0.000252,  0.000299, 0.000540])


par, cov= curve_fit(f, x, y, sigma=sigma_y, absolute_sigma=True)
print('termine noto ', par[0])
print('errore ', sqrt(cov[0,0]))
y_predetti = f(x, *par)
chi_quadro = np.sum(((y - y_predetti) / sigma_y)**2)
d=len(x)-len(par)
chi_quadro=chi_quadro/d
print('chiquadro ',chi_quadro)
x_plot=np.linspace(0.00001,4,num=100000)
plt.plot(x_plot, f(x_plot, *par), 'r', label='Risultato del fit')
plt.errorbar(x, y, yerr=sigma_y, fmt='o')
plt.xlabel(r'$a^2$')
plt.ylabel('Gap energetico')
plt.title('Estrapolazione al continuo gap energetico')
plt.xlim(0-0.01, 1+0.25)
plt.ylim(0.94, 1.01)
plt.legend()
plt.show()
```

```python
deltaEcontinuo=par[0]
deltaEcontinuoerrore=sqrt(cov[0,0])
print((deltaEcontinuo-1)/deltaEcontinuoerrore)
```
