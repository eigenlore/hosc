```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy.optimize import curve_fit
def espon(x, tau):
    return np.exp(-x/tau)
```

**t=1**

```python
gamma=np.loadtxt('../main/data_files/gamma/gamma_t1_N1024.dat')
plt.semilogy(gamma[:,0], gamma[:,1], 'r', label='')
plt.xlabel(r'$t_M$')
plt.ylabel(r'$\frac{\Gamma(t_M)}{\Gamma(0)}$')
plt.title('Funzione di autocorrelazione  '+ r'$t=1$')
plt.legend()
plt.show()
x=gamma[:,0]
y=gamma[:,1]
par, cov = curve_fit(espon, x, y)
print(par)
```
